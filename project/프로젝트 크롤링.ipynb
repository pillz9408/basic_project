```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get('https://www.youtube.com/results?search_query=%EB%8B%A4%EC%9D%B4%EC%96%B4%ED%8A%B8+%EC%98%81%EC%96%91%EC%84%B1%EB%B6%84%ED%91%9C')
```

```python
title = driver.find_elements_by_css_selector('#video-title') 
view = driver.find_elements_by_css_selector('#metadata-line > span:nth-child(1)')
```

```python
title_list =[]
for i in title:
    title_list.append(i.text)
    
view_list =[]
for i in view:
    view_list.append(i.text)
```

```python
len(title_list),len(view_list)
```

```python
view_list =[]
for i in view:
    string = i.text
    string = string.replace('조회수', '')
    string = string.replace('회', '')
    string = string.replace('.', '')
    string = string.replace('천', '')
    string = string.replace('만', '')
    string = int(string)
    view_list.append(string)
    
    print(string)
```

```python
data = {'제목' : title_list, '조회수' : view_list}
df = pd.DataFrame(data)
df.reset_index(drop=True)  
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")      # 그래프 출력 시 이상한 에러들 무시


from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = 'C:/Windows/Fonts/malgun.ttf').get_name()
rc('font',family=font_name)
```
